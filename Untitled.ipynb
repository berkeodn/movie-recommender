```python
import pandas as pd
import numpy as np

df = pd.read_csv("desktop/ratings.csv")
df.head()
```

```python
movies = pd.read_csv("desktop/movies.csv")
movies.head()
```

```python
# merging two files into one.
df = df.merge(movies, on="movieId", how="left")
df.head(10)
```

```python
# finding average ratings for each movie.

Average_ratings = pd.DataFrame(df.groupby("title")["rating"].mean())

Average_ratings.head(10)
```

```python
# number of rate for each movie is important while determining average ratings so we add new column for number of total ratings.

Average_ratings["Total Ratings"] = pd.DataFrame(df.groupby("title")["rating"].count())

Average_ratings.head(10)
```

```python
# creating pivot table
movie_user = df.pivot_table(index='userId',columns='title',values='rating')

movie_user.head(10)
```

```python
# finding correlations with the film specified.
correlations = movie_user.corrwith(movie_user["Die Hard: With a Vengeance (1995)"])
correlations.head(-10)
```

```python
#creating recommendation table by getting rid of NaN values and joining total ratings column.
rec = pd.DataFrame(correlations, columns=["Correlation"])

rec.dropna(inplace=True)

rec = rec.join(Average_ratings["Total Ratings"])

rec.head(10)
```

```python
#testing the recommendation system.
recommendation = rec[rec['Total Ratings']>100].sort_values('Correlation',ascending=False)
recommendation.head()
```

```python
#more information about highly correlated movies in order to verify results.
recommendation = recommendation.merge(movies, on="title")
recommendation.head(10)
```
